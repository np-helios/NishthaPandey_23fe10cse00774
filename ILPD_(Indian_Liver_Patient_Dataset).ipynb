```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00225/Indian%20Liver%20Patient%20Dataset%20(ILPD).csv"

df = pd.read_csv(url, header=None)

df.columns = [
    "Age", "Gender", "Total_Bilirubin", "Direct_Bilirubin",
    "Alkaline_Phosphotase", "Alamine_Aminotransferase",
    "Aspartate_Aminotransferase", "Total_Proteins",
    "Albumin", "Albumin_Globulin_Ratio", "Selector"
]

df.head()
```

```python
df['Gender'] = df['Gender'].map({
    'Male': 1,
    'Female': 0
})
```

```python
df.dtypes
```

```python
continuous_cols = [
    'Age', 'Gender', 'Total_Bilirubin', 'Direct_Bilirubin',
       'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
       'Aspartate_Aminotransferase', 'Total_Proteins', 'Albumin',
       'Albumin_Globulin_Ratio', 'Selector'
]

outlier_counts = {}

for col in continuous_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_counts[col] = len(outliers)

pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier_Count']) \
  .sort_values(by='Outlier_Count', ascending=False)
```

```python
df['Gender_log'] = np.log1p(df['Gender'])
df['Total_Bilirubin_log'] = np.log1p(df['Total_Bilirubin'])
df['Direct_Bilirubin_log'] = np.log1p(df['Direct_Bilirubin'])
df['Alkaline_Phosphotase_log'] = np.log1p(df['Alkaline_Phosphotase'])
df['Alamine_Aminotransferase_log'] = np.log1p(df['Alamine_Aminotransferase'])
```

```python
from sklearn.preprocessing import MinMaxScaler
X = df[
    [
        'Gender_log',
        'Total_Bilirubin_log',
        'Direct_Bilirubin_log',
        'Alkaline_Phosphotase_log',
        'Alamine_Aminotransferase_log'
    ]
]

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
print(X_normalized[:5])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
print(X_standardized[:5])
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X_normalized[:, 0], X_normalized[:, 1])
plt.xlabel("Total_Bilirubin_log (normalized)")
plt.ylabel("Direct_Bilirubin_log (normalized)")
plt.title("Outlier / Error Check")
plt.show()
```
